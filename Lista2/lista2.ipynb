Carrega uma planilha com dados de homicídios entre 2017 a 2022 no estado de São Paulo. Concatena a planilha de cada ano em uma.

```python
import pandas as pd

filename = 'Homicidio_2017_2022.xlsx'

file = pd.ExcelFile(filename)

ano_2017 = pd.read_excel(file, '2017')
ano_2018 = pd.read_excel(file, '2018')
ano_2019 = pd.read_excel(file, '2019')
ano_2020 = pd.read_excel(file, '2020')
ano_2021 = pd.read_excel(file, '2021')
ano_2022 = pd.read_excel(file, '2022')

#data = pd.merge(ano_2017, ano_2018, ano_2019, ano_2020, ano_2021, ano_2022, how = 'outer')
data = pd.concat([ano_2017, ano_2018, ano_2019, ano_2020, ano_2021, ano_2022])
data
```

Tira as linhas duplicadas.

```python
data = data.drop_duplicates()
```

Coloca todos os dados da planilha em low case.

```python
for i in data.columns:
  data[i]=data[i].astype(str).str.lower()

data
```

```python
Filtra os dados para a cidade de São Paulo.
```

```python
filtro1 = data['MUNICIPIO_CIRCUNSCRICAO'] == 'são paulo'

data[filtro1]
```

Faz uma cópia do filtro anterior.

```python
datac = data[filtro1].copy()
```

Transforma os dados númerios de object para int64.

```python
datac = datac.applymap(lambda x: int(x) if str(x).isnumeric() else x )
```

```python
datac.dtypes
```

```python
datac
```

Tira o caractere ª da coluna 'SECCIONAL_CIRCUNSCRICAO'.

```python
datac['SECCIONAL_CIRCUNSCRICAO'] = datac['SECCIONAL_CIRCUNSCRICAO'].str.replace('ª','')
datac
```

Tira o caractere º da coluna 'DP_ELABORACAO'.

```python
datac['DP_ELABORACAO'] = datac['DP_ELABORACAO'].str.replace('º','')
datac
```

Tira o caractere º 'SEC_ELABORACAO'

```python
datac['SEC_ELABORACAO'] = datac['SEC_ELABORACAO'].str.replace('º','')
datac
```

Coloca 'nan' nos dados faltantes na coluna 'HD'

```python
datac['HD'] = datac['HD'].str.replace(' ','nan')
datac
```

Transfere os valores da coluna 'DATAHORA_REGISTRO_BO' para um vetor.

```python
vetor = datac['DATAHORA_REGISTRO_BO'].values
vetor
```

Filtra a data.

```python
data = []
for i in vetor:
  data.append(i.split(' ')[0])
```

Divide o mês e o dia.

```python
mes = []
dia = []
for i in data:
  mes.append(int(i.split('-')[1]))
  dia.append(int(i.split('-')[2]))
```

```python
datac1 = datac.drop(['DATAHORA_REGISTRO_BO'], axis=1)
datac1
```

Cria novas colunas com mês e dia.

```python
datac1['MES_BO'] = mes
datac1['DIA_BO'] = dia

datac1
```

Alguns dados númeris na coluna 'IDADE_PESSOA' estavam como object e com '.0'. Essa função lambda transforma para int64 e tira o '.0'.

```python
datac1['IDADE_PESSOA'] = datac1['IDADE_PESSOA'].map(lambda x: int(x.replace('.0', '')) if str(x).find('.0') != -1 else( int(x) if str(x).isnumeric() else x ))
datac1
```

Transforma dos dados númericos de 'LATITUDE' e 'LONGITUDE' em ponto flutuante.

```python
datac1['LATITUDE'] = datac1['LATITUDE'].map(lambda x: float(x) if str(x).isnumeric() else x)
datac1
```

```python
datac1['LONGITUDE'] = datac1['LONGITUDE'].map(lambda x: float(x) if str(x).isnumeric() else x)
datac1
```

Filtra o número de homícidios em cada ano e transforma os dados em um gráfico de barras.

```python
list_ano = [2017, 2018, 2019, 2020, 2021, 2022]
list_per = []
for i in list_ano:

  filtro_ano = datac1['ANO_BO'] == i

  list_per.append(len([x for x in filtro_ano if x == True]))

list_per
```

```python
import matplotlib.pyplot as plt

width = 0.35
fig, ax = plt.subplots()

pps = ax.bar(list_ano, list_per, width, align='center')

for p in pps:
   height = p.get_height()
   ax.text(x=p.get_x() + p.get_width() / 2, y=height+.10,
      s="{}".format(height),
      ha='center')

plt.show()
```

Filtra os dados da cor das vítimas e constrói um gráfico com os dados.

```python
list_cor = ['branca', 'preta', 'parda']
list_per = []

for i in list_cor:

  filtro_cor = datac1['COR_PELE'] == i

  list_per.append(len([x for x in filtro_cor if x == True]))

list_cor.append('outros')

list_per.append(4048-(sum(list_per)))
```

```python
# Dados para plotar


# definições
colors = ['lightcoral','gold', 'yellowgreen',  'lightskyblue']

# explodir 1ª fatia
explode = (0, 0, 0, 0.1)  

# Plot
plt.pie(list_per, explode=explode, labels=list_cor, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

plt.legend(list_cor, loc="lower right")

plt.axis('equal')
plt.show()
```
