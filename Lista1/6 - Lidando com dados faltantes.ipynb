## Identificando valores faltantes em dados tabulares
Vamos iniciar nosso estudo identificando dados faltantes em tabelas.

Carrega um arquivo csv em df.

```python
import pandas as pd

df = pd.read_csv('missing.csv')
df
```

Conta quantos valores estão faltando em cada coluna.

```python
df.isnull().sum()
```

Cria um array com os valores do df.

```python
# Cria um array numpy com todos os valores do DataFrame

df.values
```

## Eliminando amostras ou *features* com valores ausentes

Remove as linhas que faltam valores.

```python
# Remove colunas que contém valores faltantes

df.dropna(axis=0)
```

Remove colunas que faltam valores.

```python
# Removendo colunas que contém valores faltantes

df.dropna(axis=1)
```

Remove todas as linhas onde todas as colunas faltam valores.

```python
# "Drop" apenas nas linhas onde todas as colunas são NaN

df.dropna(how='all')
```

Remove as linhas que possuem menos de 4 valores reais.

```python
# Removendo linhas que possuem menos de 3 valores reais

df.dropna(thresh=4)
```

Remove as linhas onde faltam valores em uma coluna específica.

```python
# Remove apenas linhas onde NaN aparece em uma coluna específica

df.dropna(subset=['C'])
```
