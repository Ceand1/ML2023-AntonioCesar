# 3 - Tipos e Conversão de tipos
O Numpy suporta uma variedade de tipos muito maior do que o Python.

Existem 5 tipos básicos numéricos
- Booleans (bool)
- Integers (int)
- Unsigned Intergers (uint)
- Floating Points (float)
- Complex

Complementarmente, a célula abaixo mostra um quadro comparativos com uma maior quantidade de tipos do Numpy.

```python
import numpy as np
```

x recebe 1.0 como um float de 32 bits.

```python
x = np.float32(1.0)
x
```

y recebe um array de inteiros.

```python
y = np.int_([1,2,4])
y
```

Foi especificado apenas o valor de parada, então os valores gerados serão de 0 a 3, inteiros sem sinal de 8 bits.

```python
z = np.arange(3, dtype=np.uint8)
z
```

Converte uma lista de três elementos inteiros para um array de três elementos do tipo float.

```python
np.array([1, 2, 3], dtype='f')
```

## Convertendo tipos de arrays

Para convertermos arrays Numpy, basta utilizarmos o método astype().

Converte z para um array do tipo float.

```python
z.astype(float)
```

Converte o array z para tipo inteiro de 8 bits.

```python
np.int8(z)
```

Para verificarmos o tipo de um array Numpy, basta verificarmos o seu atributo dtype.

```python
z.dtype
```
