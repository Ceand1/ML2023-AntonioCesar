## Operadores booleanos

### and

Retorna true se x for maior que 5 e menor que 15.

```python
x = 12
```

```python
x > 5 and x < 15
```

### or

Retorna true se y for menor que 7 ou maior que 13.

```python
y = 5
```

```python
y < 7 or y > 13
```

### not

Nega a expressão.

```python
not True
```

```python
not False
```

### Numpy

```python
import numpy as np

np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
bmi = np_weight / np_height ** 2

bmi
```

Para cada elemento do array retorna true se ele for mais que 21 e false se for menor ou igual a 21.

```python
bmi > 21
```

```python
bmi > 21 and bmi < 22
```

Para todos os elementos do array, retorna true se o valor for entre 21 e 22.

```python
np.logical_and(bmi > 21, bmi < 22)
```

Retorna apenas os valores que estão entre 21 e 22 no array.

```python
bmi[np.logical_and(bmi > 21, bmi < 22)]
```

## Filtrando um DataFrame Pandas

```python
import pandas as pd
brics = pd.read_csv('brics.csv', index_col=0)
brics
```

**Objetivo**: selecionar os países com área maior do que 8 milhões de km2

3 passos:
- Selecionar a coluna "area"
- Fazer uma comparação na coluna "area"
- Usar o resultado para selecionar os países

```python
# Selecionar a coluna "area"
brics['area']
```

```python
# Fazer uma comparação na coluna "area"
brics['area'] > 8
is_huge = brics['area'] > 8
is_huge
```

```python
# Usar o resultado para selecionar os países
brics[is_huge]
```

Ler um arquivo csv.

```python
# Importando dataset de carros
import pandas as pd
cars = pd.read_csv('cars.csv', index_col = 0)
cars
```

Seleciona uma coluna específica e cria um subconjunto apenas com os dados dessa coluna.

```python
# Extraia a coluna 'drives_right' do DataFrame 'cars' e atribua a variável 'dr'
dr = cars['drives_right']

# Use a variável 'dr' para criar um subconjunto de dados chamado 'sel'
sel = cars[dr]

# Imprima a variável 'sel'
print(sel)
```

Car_maniac recebe apenas as linhas que possuirem valores na coluna cpc maiores que 500.

```python
# Crie uma variável car_maniac: linhas que possuem 'cars_per_cap' maiores do que 500
cpc = cars['cars_per_cap']
many_cars = cpc > 500
car_maniac = cars[many_cars]

# Imprima 'car_maniac'
print(car_maniac)
```

Retorna apenas as linhas onde a coluna cpc possui valor entre 100 e 500.

```python
# Importe o conjunto de dados 'cars.csv', bem como as bibliotecas 'pandas' e 'numpy'
import pandas as pd
import numpy as np
cars = pd.read_csv('cars.csv', index_col = 0)

# Crie a variável 'medium': linhas que possuem 'cars_per_cap' entre 100 e 500
cpc = cars['cars_per_cap']
between = np.logical_and(cpc > 100, cpc < 500)
medium = cars[between]

# Imprima 'medium'
print(medium)
```
